Visualisation d’un nuage de points et exploration des
variables

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

pd.set_option('display.max_columns', 100)
```

```python
house = pd.read_csv('./files/house_pricing.csv', ','); house
```

```python
house.info()
```

```python
sb.relplot(x="SalePrice", y="LotArea",  hue="SalePrice", data=house)
```

```python
house_500 = house[house['SalePrice'] < 500000]
house_2 = house_500[house_500['LotArea'] < 20000]
sb.relplot(x="SalePrice", y="LotArea",  hue="SalePrice", data=house_2)
```

```python
sb.relplot(x="LotFrontage", y="LotArea", hue="LotFrontage", data = house)
```

```python
house_200 = house[house['LotFrontage'] < 200]
house_2 = house_200[house_200['LotArea'] < 100000]
sb.relplot(x="LotFrontage", y="LotArea", hue="LotFrontage", data = house_2)
```
