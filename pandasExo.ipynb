```python
ibm = pd.read_csv('./files/IBM_HR_ATTRITION.csv', ','); 
pd.set_option('display.max_rows', 100)
ibm
```

```python
ibm.describe(include="all")
```

```python
sb.catplot(y="JobSatisfaction", x="Attrition", data=ibm, kind="box")
```

```python
sb.catplot(x="JobSatisfaction", data = ibm[ibm['Attrition'] == "Yes"], kind="count")
```

```python
sb.boxplot(x=ibm[ibm['Attrition'] == "Yes"]['JobSatisfaction'])
```

```python
sb.catplot(y="TotalWorkingYears", x="Attrition", data=ibm, kind="box")
```

```python
ibm[ibm['Attrition'] == "Yes"]['TotalWorkingYears'].describe() 
```

```python
q1y = ibm[ibm['Attrition'] == "Yes"]['TotalWorkingYears'].quantile(.25)
q3y = ibm[ibm['Attrition'] == "Yes"]['TotalWorkingYears'].quantile(.75)
q1n = ibm[ibm['Attrition'] == "No"]['TotalWorkingYears'].quantile(.25)
q3n = ibm[ibm['Attrition'] == "No"]['TotalWorkingYears'].quantile(.75)
```

```python
ibm_yes = ibm[(ibm['TotalWorkingYears'] > q3y+(1.5*(q3y-q1y))) & (ibm['Attrition'] == "Yes")]; ibm_yes
```

```python
ibm_no = ibm[(ibm['TotalWorkingYears'] > q3n+(1.5*(q3n-q1n))) & (ibm['Attrition'] == "No")]; ibm_no
```

```python
ibm_iso = ibm[(ibm['TotalWorkingYears'] < q3y+(1.5*(q3y-q1y)))]
```

```python
sb.catplot(y="TotalWorkingYears", x="Attrition", data=ibm_iso, kind="box")
```

```python
ibm_yes.describe()
```

Visualisation dâ€™une heatmap

```python
pricing = pd.read_csv('./files/house_pricing.csv')
house = pricing.iloc[:, -15:]; house
```

```python
corr = house.corr(); corr
```

```python
sb.heatmap(corr)
```

```python
f, ax = plt.subplots(figsize=(20,5))
sb.heatmap(corr, annot=True, ax =ax)
```

```python
house_2 = pricing.iloc[:, [3,4,80]]; house_2
```

```python
corr_2 = house_2.corr(); corr_2
```

```python
f, ax = plt.subplots(figsize=(20,5))
sb.heatmap(corr_2, annot=True, ax =ax)
```
